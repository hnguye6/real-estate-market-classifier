# Data Exploration & Manipulation

## Load all the tables

```python
# Import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load all the tables
share_df = pd.read_csv("share_of_listings_price_cut.csv")
market_df = pd.read_csv("market_heat_index.csv") # Primary table
mean_df = pd.read_csv("mean_days_to_pending.csv")
price_df = pd.read_csv("price_tier_context.csv")
salelist_df = pd.read_csv("sale_to_list_ratio.csv")
```

## Transform the tables

 I melted all five tables from wide format (one column per date) into long format (one row per metro-date combination), with a Date and variable column (e.g PriceCutShare for share_df) replacing the 100+ separate date columns.

```python
# Transform the share_df
share_df_new = pd.melt(
    share_df,
    id_vars=['RegionID', 'SizeRank',	'RegionName',	'RegionType', 'StateName'],
    var_name='Date',
    value_name='PriceCutShare'
)

# Convert Date from string to actual datetime — don't leave this as object dtype
share_df_new['Date'] = pd.to_datetime(share_df_new['Date'])
```

```python
# Transform the market_df
market_df_new = pd.melt(
    market_df,
    id_vars=['RegionID', 'SizeRank',	'RegionName',	'RegionType', 'StateName'],
    var_name='Date',
    value_name='MarketHeatIndex'
)

# Convert Date from string to actual datetime — don't leave this as object dtype
market_df_new['Date'] = pd.to_datetime(market_df_new['Date'])
```

```python
# Transform the mean_df
mean_df_new = pd.melt(
    mean_df,
    id_vars=['RegionID', 'SizeRank',	'RegionName',	'RegionType', 'StateName'],
    var_name='Date',
    value_name='MeanDaysToPending'
)

# Convert Date from string to actual datetime — don't leave this as object dtype
mean_df_new['Date'] = pd.to_datetime(mean_df_new['Date'])
```

```python
# Transform the price_df
price_df_new = pd.melt(
    price_df,
    id_vars=['RegionID', 'SizeRank',	'RegionName',	'RegionType', 'StateName'],
    var_name='Date',
    value_name='ZHVI'
)

# Convert Date from string to actual datetime — don't leave this as object dtype
price_df_new['Date'] = pd.to_datetime(price_df_new['Date'])
```

```python
# Transform the salelist_df
salelist_df_new = pd.melt(
    salelist_df,
    id_vars=['RegionID', 'SizeRank',	'RegionName',	'RegionType', 'StateName'],
    var_name='Date',
    value_name='SaleToListRatio'
)

# Convert Date from string to actual datetime — don't leave this as object dtype
salelist_df_new['Date'] = pd.to_datetime(salelist_df_new['Date'])
```

```python
new_tables =  {
    "share_df_new": share_df_new,
    "market_df_new": market_df_new, # Primary table
    "mean_df_new": mean_df_new,
    "price_df_new": price_df_new,
    "salelist_df_new": salelist_df_new,
}
for name, df in new_tables.items():
  print(f'{"="*120}')
  print(f'{name} table')
  print(f'{"="*120}')
  display(df.head(3))
  print('')
```

```python
for name, df in new_tables.items():
  print(f'{"="*60}')
  print(f'{name} table')
  print(f'{"="*60}')
  display(df.info())
  print("")
```

## Set up SQL

```python
import sqlite3

conn = sqlite3.connect(':memory:')

# Load all dataframes into SQLite
share_df_new.to_sql('share',conn,if_exists='replace',index=False)
market_df_new.to_sql('market',conn,if_exists='replace',index=False) # Primary table
mean_df_new.to_sql('mean',conn,if_exists='replace',index=False)
price_df_new.to_sql('price',conn,if_exists='replace',index=False)
salelist_df_new.to_sql('salelist',conn,if_exists='replace',index=False)
```

```python
def run_sql(sql):
  """Run a SQL query and return results as a pandas DataFrame."""
  return pd.read_sql(sql, conn)
```

```python
# Test SQL in Google Colab
run_sql("""
    SELECT *
    FROM mean
    LIMIT 3
""")
```

## Merge all tables

```python
master_raw = run_sql("""
  SELECT
    m.RegionID,
    m.SizeRank,
    m.RegionName,
    m.RegionType,
    m.StateName,
    m.Date,
    sh.PriceCutShare,
    m.MarketHeatIndex,
    a.MeanDaysToPending,
    p.ZHVI,
    sl.SaleToListRatio
  FROM market AS m
  LEFT JOIN share AS sh
    ON m.RegionID = sh.RegionID AND m.Date = sh.Date
  LEFT JOIN mean AS a
    ON m.RegionID = a.RegionID AND m.Date = a.Date
  LEFT JOIN price AS p
    ON m.RegionID = p.RegionID AND m.Date = p.Date
  LEFT JOIN salelist AS sl
    ON m.RegionID = sl.RegionID AND m.Date = sl.Date
""")
```

### PriceCutShare and MeanDaysToPending

```python
# Check missingness by metro size
master_raw['SaleToListRatio_missing'] = master_raw['SaleToListRatio'].isnull()
master_raw['MeanDaysToPending_missing'] = master_raw['MeanDaysToPending'].isnull()

print("SaleToListRatio missing rate by size quartile:")
print(master_raw.groupby(pd.qcut(master_raw['SizeRank'], 4))['SaleToListRatio_missing'].mean())

print("\nMeanDaysToPending missing rate by size quartile:")
print(master_raw.groupby(pd.qcut(master_raw['SizeRank'], 4))['MeanDaysToPending_missing'].mean())
```

```python
# Check missingness over time (by year)
master_raw['Year'] = master_raw['Date'].dt.year

print("SaleToListRatio missing rate by year:")
print(master_raw.groupby('Year')['SaleToListRatio_missing'].mean())

print("\nMeanDaysToPending missing rate by year:")
print(master_raw.groupby('Year')['MeanDaysToPending_missing'].mean())
```

```python
master_raw = master_raw.drop(columns=['PriceCutShare','MeanDaysToPending'])
```

**Why PriceCutShare and PriceCutShare were dropped?**

PriceCutShare is one of the three direct inputs Zillow uses to calculate the Market Heat Index, so including it as a "predictor" would partly mean reconstructing a known ingredient of the label rather than learning genuine signal.

MeanDaysToPending was dropped for two separate reasons: it belongs to the same underlying "days to pending" data as another of the index's formula inputs (a similar leakage risk), and separately it has severe, non-random missingness (48%) concentrated in smaller metros. Dropping both keeps the classifier honestly predictive and avoids fabricating data for the market segments most relevant to an investment firm's decisions.

```python
master_raw = master_raw.dropna(subset=['SaleToListRatio'])
```

**Why rows with missing SaleToListRatio were dropped?**

The missing values weren't random — they were concentrated almost entirely in smaller, lower-liquidity metros (up to 98% missing in the smallest quartile), since Zillow needs a minimum sales volume to reliably calculate the ratio. Filling these gaps with an imputed value like a median would mean inventing data for exactly the smaller markets an investment firm might care about most. Dropping these rows keeps every remaining data point genuine, at the honest cost of reduced metro coverage — a limitation clearly noted rather than hidden behind fabricated numbers.
